## Q11) 

```python
#Importing required modules
from scipy import stats

s_mean = 200
s_size = 2000
DoF = 1999
s_std = 30
#Formula for Confidence Interval = x_bar +/- t(1-alpha/2, n-1)* s/ (root of n)

alpha_values = [0.97, 0.99, 0.98]
# alpha = 100% - given percentage ex : 94%, 98%, 96% = 0.06, 0.02, 0.04 respectively.
t_values = [stats.t.ppf(alpha, DoF) for alpha in alpha_values]
# By using stats .t.ppf we can get t value at required alpha value.

# Formula for confidence interval.
confidence_intervals = [( s_mean - t * s_std / (s_size ** 0.5), s_mean + t * s_std / (s_size ** 0.5)) for t in t_values ]

print(f"94% confidence interval: {confidence_intervals[0]}")
print(f"98% confidence interval: {confidence_intervals[1]}")
print(f"96% confidence interval: {confidence_intervals[2]}")
```

```python
from scipy.stats import norm

# Calculate the Z score for a 90% confidence interval
z_90 = norm.ppf(0.95)

# Calculate the Z score for a 94% confidence interval
z_94 = norm.ppf(0.97)

# Calculate the Z score for a 60% confidence interval
z_60 = norm.ppf(0.8)

print(f"The Z score for a 90% confidence interval is {z_90:.2f}.")
print(f"The Z score for a 94% confidence interval is {z_94:.2f}.")
print(f"The Z score for a 60% confidence interval is {z_60:.2f}.")
```
