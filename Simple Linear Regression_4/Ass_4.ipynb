### <font color = "white">Delivery_time -> Predicting delivery time using sorting time 

```python
# Import required modules
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Read the csv file into data_1
data_1 = pd.read_csv("delivery_time.csv")
#print(data_2.head())

# check the data_2_2_2_2_2 in order to perform EDA steps or not ?
print(data_1.info())

# Check the correlation 
print(data_1.corr())

# Read the data into separate columns
x = data_1['Sorting Time']
y = data_1['Delivery Time']

# Perform OLS method to get b0 and b1 values in order to fit the line 
model_1 = smf.ols("y~x", data = data_1).fit()
#  Use summary() function to get our model details
print(model_1.summary())

# Values of b0 and b1 from model.summary()
b0 = 6.5827 
b1 = 1.6490 
# Line equation
y_hat = b0 + b1*x

# Ploting line plot and scatter plot for our visulization
plt.plot(x, y_hat)
plt.scatter(x, y, c="black")

# Labeling the x and y axis
plt.xlabel("Sorting Time")
plt.ylabel("Delivery Time")
```

```python
# New sorting time
new_sorting_time = 8 # replace with your value

# Create a new DataFrame for the new sorting time
new_data = pd.DataFrame({'x': [new_sorting_time]})

# Use the model to predict the delivery time for the new sorting time
new_delivery_time = model_1.predict(new_data)

print(f"The predicted delivery time for a sorting time of {new_sorting_time} is {new_delivery_time[0]}")
```

### <font color = "white">Salary_hike -> Building a prediction model for Salary_hike

```python
# Import required modules
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Read the csv file into data
data_2 = pd.read_csv("Salary_Data.csv")
#print(data.head())

# check the data in order to perform EDA steps or not ? 
print(data_2.info())

# Check the correlation 
print(data_2.corr())

# Read the data into separate columns
x = data_2['YearsExperience']
y = data_2['Salary']

# Perform OLS method to get b0 and b1 values in order to fit the line 
model_2 = smf.ols("y ~ x", data = data_2).fit()
#  Use summary() function to get our model details
print(model_2.summary())

# Values of b0 and b1 from model.summary()
b0 = 2.579e+04  
b1 = 9449.9623
# Line equation
y_hat = b0 + b1*x

# Ploting line plot and scatter plot for our visulization
plt.plot(x, y_hat)
plt.scatter(x, y, c="black")

# Labeling the x and y axis
plt.xlabel("Years Experience")
plt.ylabel("Salary")
```

```python
# Year of experience
Years_experience = 6 # replace with your value

# Create a new DataFrame for the Year of Experience
new_data = pd.DataFrame({'x': [Years_experience]})

# Use the model to predict the Salary hike for the new year of experience
Salary_hike = model_2.predict(new_data)

print(f"The predicted salary hike for year of experience of {Years_experience} is {Salary_hike[0]}")
```
